```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.initializers import he_normal
from matplotlib import pyplot as plt
import os
```

```python
train_dir=r"F:\Files\ANN\archive\Train\Train"
train_data_gen= ImageDataGenerator(rescale=(1./255),
                                    rotation_range=40,
                                    width_shift_range=0.2,
                                    height_shift_range=0.2,
                                    shear_range=0.2,
                                    zoom_range=0.2,
                                    # horizontal_flipd=True,
                                    fill_mode="nearest"
)
train_generator = train_data_gen.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical",
    shuffle=True
)
```

```python
val_dir=r"F:\Files\ANN\archive\Validation\Validation"
val_data_gen= ImageDataGenerator(rescale=(1./255))
val_generator = val_data_gen.flow_from_directory(
    val_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical",
    shuffle=True
)
```

```python
test_dir=r"F:\Files\ANN\archive\Test\Test"
test_data_gen= ImageDataGenerator(rescale=(1./255))
test_generator = test_data_gen.flow_from_directory(
    test_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical",
    shuffle=True
)
```

```python
print("Training images of class Healthy: ", len(os.listdir(os.path.join(train_dir,"Healthy"))))
print("Training images of class Powdery: ", len(os.listdir(os.path.join(train_dir,"Powdery"))))
print("Training images of class Rust: ", len(os.listdir(os.path.join(train_dir,"Rust"))))
```

```python
model = Sequential([
    Conv2D(filters=95,kernel_size=(11,11), strides=4,input_shape=(224,224,3),activation="relu"),
    MaxPooling2D(pool_size=(3,3),strides=2),
    Conv2D(filters=256,kernel_size=(5,5), padding="same",input_shape=(224,224,3),activation="relu"),
    MaxPooling2D(pool_size=(3,3),strides=2),
    Conv2D(filters=384,kernel_size=(3,3), padding="same",input_shape=(224,224,3),activation="relu"),
    Conv2D(filters=384,kernel_size=(3,3), padding="same",input_shape=(224,224,3),activation="relu"),
    Conv2D(filters=256,kernel_size=(3,3), padding="same",input_shape=(224,224,3),activation="relu"),
    MaxPooling2D(pool_size=(3,3),strides=2),
    Flatten(),
    Dense(500,activation="relu"),
    Dropout(0.5),
    Dense(324,activation="relu"),
    Dropout(0.5),
    Dense(3,activation="softmax")
])
```

```python
model.compile(loss="categorical_crossentropy",
               optimizer=Adam(learning_rate=0.0001),
               metrics=["acc"]
)
```

```python
model.fit(train_generator,
        epochs=10,
        batch_size=128,
        validation_data=val_generator,
        verbose=1
)
```
